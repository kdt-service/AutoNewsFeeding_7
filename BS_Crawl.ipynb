# 마지막 페이지 해결 Ver
- url 갖고오기 : 목록에서 li 태그들이 기사를 의미. li 태그에서 a태그로 들어가 href를 통해 갖고 올수있다. 
- 마지막 페이지는 기사의 개수가 15개보다 작거나 같다. 
- - 기사의 개수가 15개보다 작다면 -> 거기는 마지막 페이지 
- - 그 다음페이지로 넘겼는데 -> li tag가 없어. 그럼 그 전 페이지가 마지막 페이지.

```python
URL = 'https://news.daum.net/breakingnews/society?page={}&regDate={}'
he = {"user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36"}
response = requests.get(URL, headers = he)
# 얻은 정보들은 news_df에 저장
news_df = pd.DataFrame(columns = ['date','page','title','press','context'])

# date에서 일자 지정 가능, page에서 페이지 지정 가능 -> 각각 range로 범위 지정 가능
warn  = 0 
for date in range(20230303, 20230304):
    for page in range(400,1000):
        response = requests.get(URL.format(page,date), headers = he)
        soup = BeautifulSoup(response.text, 'html.parser')
        li_tags = soup.select(".list_news2.list_allnews li")
        for li in li_tags:
            title = li.select_one('.cont_thumb .tit_thumb .link_txt').text.strip() # 제목
            press = li.select_one('.cont_thumb .tit_thumb .info_news').text.strip()[:-8] # 언론사
            context = li.select_one('.cont_thumb .desc_thumb .link_txt').text.strip() # 내용
             # 마지막 페이지 2차 확인
            if len(news_df)>=15 :
                if news_df.loc[len(news_df)-15].context == context :
                    warn +=1
            if warn == 0 :                
                news_df.loc[len(news_df)]=[date, page, title, press, context] 
        if warn > 0:
            break
        # 마지막 페이지 1차 확인 -> 더이상 page 넘기지 말고 날짜를 넘겨! 
        if len(li_tags)<15 :
            break
       
```

# 합치기 Ver

```python
URL = 'https://news.daum.net/breakingnews/society?page={}&regDate={}'
he = {"user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36"}
response = requests.get(URL, headers = he)
# 얻은 정보들은 news_df에 저장
news_df = pd.DataFrame(columns = ['date','page','title','press','context','url'])

# date에서 일자 지정 가능, page에서 페이지 지정 가능 -> 각각 range로 범위 지정 가능
warn  = 0 
for date in range(20230303, 20230304):
    for page in range(400,1000):
        response = requests.get(URL.format(page,date), headers = he)
        soup = BeautifulSoup(response.text, 'html.parser')
        li_tags = soup.select(".list_news2.list_allnews li")
        
        for li in li_tags:
            title = li.select_one('.cont_thumb .tit_thumb .link_txt').text.strip() # 제목
            press = li.select_one('.cont_thumb .tit_thumb .info_news').text.strip()[:-8] # 언론사
            context = li.select_one('.cont_thumb .desc_thumb .link_txt').text.strip() # 잘린내용
            url= li.find("a", {'class':'link_txt'}).get('href')
            #url 통해 내용 가져오기 부분! 
            url_response=requests.get(url, headers = he)
            url_soup= BeautifulSoup(url_response.text, 'html.parser')
            generals = url_soup.select('section p[dmcf-ptype=general]')
            real_text = ""
            for general in generals:
                real_text+=general.text.strip()
                
             # 마지막 페이지 2차 확인
            if len(news_df)>=15 :
                if news_df.loc[len(news_df)-15].context == context :
                    warn +=1
            if warn == 0 : 
                # 데이터 넣는 부분
                news_df.loc[len(news_df)]=[date, page, title, press, real_text,url] 
        if warn > 0:
            break
        # 마지막 페이지 1차 확인 -> 더이상 page 넘기지 말고 날짜를 넘겨! 
        if len(li_tags)<15 :
            break
       
```
